```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
customer_data = pd.read_csv('shopping-data.csv')

customer_data.head()
```

```python
customer_data.describe()
```

```python
# Bivariate analysis

# Scatter plot
plt.scatter(customer_data.iloc[:,2].values, customer_data.iloc[:,3].values)
plt.xlabel("Age", size=12)
plt.ylabel("Annual income", size=12)
plt.show()
```

```python
# Scatter plot 2

customer_data.plot.scatter(x="Genre", y="Annual Income (k$)")
plt.show
```

```python
# Pair plot using seaborn library

sns.pairplot(data=customer_data, vars=['Age','Annual Income (k$)','Spending Score (1-100)'])
plt.show()
```

```python
# Correlation matrix
customer_data[['Age','Annual Income (k$)','Spending Score (1-100)']].corr()

sns.heatmap(customer_data[['Age','Annual Income (k$)','Spending Score (1-100)']].corr(), annot=True, cmap='Reds')
plt.show
```

```python
sns.boxplot(data=customer_data, x="Annual Income (k$)", y="Genre")
plt.show
```

```python
# Comparing numerical and categorical data

customer_data.groupby('Genre')['Annual Income (k$)'].mean()
```
